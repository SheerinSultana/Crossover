```python
important_columns = ['Age', 'Gender', 'Location', 'Income', 'Interests',
                     'Last_Login_Days_Ago', 'Purchase_Frequency', 'Average_Order_Value',
                     'Total_Spending', 'Product_Category_Preference',
                     'Time_Spent_on_Site_Minutes', 'Pages_Viewed',
                     'Newsletter_Subscription']

data= data[important_columns]
```

```python
data_features= data.drop(columns=['Product_Category_Preference'])
data_predict = data['Product_Category_Preference']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(stop_words='english')
columns_to_combine = list(data.columns)
data['Combined_Features'] = data[columns_to_combine].astype(str).apply(' '.join, axis=1)
```

```python
from sklearn.metrics.pairwise import linear_kernel
from sklearn.metrics import accuracy_score

tfidf_matrix = tfidf_vectorizer.fit_transform(data['Combined_Features'])

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Function to recommend categories based on multiple features
def recommend_categories(user_id, cosine_sim=cosine_sim):
    user_index = user_id - 1  # Adjusting user ID to match index
    sim_scores = list(enumerate(cosine_sim[user_index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Exclude the user's own index
    category_indices = [i[0] for i in sim_scores]
    return data['Product_Category_Preference'].iloc[category_indices]

# Example usage: Recommend categories for user ID 1
user_id = 1
recommended_categories = recommend_categories(user_id)
print("Recommended Categories for User", user_id, ":", recommended_categories)
```
